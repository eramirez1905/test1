## Differences in Hurrier Small Tables

```python
def get_diff_small_tables(log_dwh_all, pan_dwh_ext, pan_dwh_int, diff_threshold=0):
    # Union all data from External and Internal in Pandora DWH
    pandora_data = pd.concat([pan_dwh_ext, pan_dwh_int], ignore_index=True)
    # Merge with Left Join with Logistics dataframe
    pandora_merge = pd.merge(pandora_data,
                             log_dwh_all[['country_code', 'qty']],
                             on = ['country_code'],
                             suffixes=('_pandora', '_logistics')).dropna()
    # Sorting
    pandora_merge.sort_values("country_code", inplace=True)
    # Include diff column
    pandora_merge['qty_diff'] = pandora_merge['qty_pandora'] - pandora_merge['qty_logistics']
    
    # All differences by some threshold
    return pandora_merge[pandora_merge['qty_diff'] > diff_threshold]
```

```python
print ('Merge Layer Hurrier businesses')
get_diff_small_tables(df_log_dwh_hurrier_businesses, df_pan_ext_businesses, df_pan_int_businesses)
```

```python
print ('Merge Layer Hurrier cities')
get_diff_small_tables(df_log_dwh_hurrier_cities, df_pan_ext_cities, df_pan_int_cities)
```

```python
print ('Merge Layer Hurrier courier_feedbacks')
get_diff_small_tables(df_log_dwh_hurrier_courier_feedbacks, df_pan_ext_courier_feedbacks, df_pan_int_courier_feedbacks)
```
